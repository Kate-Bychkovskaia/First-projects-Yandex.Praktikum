```python
# new catalogue for splited files
output_directory = '/Users/ekaterina/Documents/thumos_test'
os.makedirs(output_directory, exist_ok=True)
```

```python
# Split DataFrame on rows and save each row in a new csv file
for index, row in df.iterrows():
    #file's name formatting 
    output_file_path = os.path.join(output_directory, f'data_{index + 1:012d}.csv')
    
# create DataFrame with 1 row and save as new csv
    pd.DataFrame(row).T.to_csv(output_file_path, index=False)
```

### 2. Merging files into a single dataset

```python
# directory path where CSV files will be stored
directory_path = '/Users/ekaterina/Documents/thumos_test'

# Initialize an empty DataFrame to store the merged data
merged_data = []



# Loop through all files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory_path, filename)
        # Read each CSV file and append it to the merged_data DataFrame
        df = pd.read_csv(file_path)
        merged_data.append(df)
        

#Concatenate all DataFrames in the list into a single DataFrame
merged_data = pd.concat(merged_data, ignore_index=True)

#Display the first few rows of the merged dataset
merged_data.head()
```

```python
#check the length of the final DataFrame
print("Length of the final dataset:", len(merged_data))
```
